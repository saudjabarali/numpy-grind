```python
#NP9-Replace all even numbers in an array with 0
arr=np.array([1,2,3,4,5,6,7])
print(arr)
arr[arr%2==0]=0
print(arr)
```

```python
#NP10-Extract elements greater than 25
mater25=np.arange(15,39).reshape(6,4)
print(mater25)
print(mater25[mater25>25])
```

```python
#🔵 LEVEL 3 – Operations NP11 Create two arrays and perform:Addition Subtraction Multiplication
arr1=np.arange(1,10).reshape(3,3)
arr2=np.arange(11,20).reshape(3,3)
print(arr1)
print(arr2)
print("Add",arr1+arr2)
print("sub", arr2-arr1)
print("Mul",arr1*arr2)
```

```python
#NP12 Find: Mean Median Standard deviation
print(arr1)
meaaan=np.mean(arr1)
mediann=np.median(arr1)
stddev=np.std(arr1)
print("mean=",meaaan)
print("median=",mediann)
print("stddev=",stddev)
```

```python
#NP13-Find max and min values in array
print(arr2)
minimum=np.min(arr2)
maximum=np.max(arr2)
print("min=",minimum)
print("max=",maximum)
```

```python
#NP15 Reshape array of size 12 into 3×4
arrrs=np.arange(1,13)
print(arrrs)
arrrs=np.reshape(arrrs,(3,4) )
print(arrrs)
```

```python
#NP16-Create random array of size 10 and sort it
arrrand=np.random.randint(1,100,10)
print(arrrand)
print(np.sort(arrrand))
```

```python
#Normalize an array (0–1 scaling)
normalized=(arrrand-np.min(arrrand)/np.max(arrrand)-np.min(arrrand))
print(normalized)
```

```python
#NP18-Find unique elements in array
unique=np.unique(arrrand)
print(unique)
```
